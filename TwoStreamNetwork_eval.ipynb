```python
from keras.callbacks import TensorBoard, ModelCheckpoint, EarlyStopping, CSVLogger
from TwoStreamNetwork.fuse.validModel import ResearchModels
from TwoStreamNetwork.fuse.validData import DataSet
import time
import os.path
from os import makedirs
from DataSetPathCall import UCF101_PathCall

ucf101 = UCF101_PathCall()

print(ucf101.getDataListPath())
```

```python
saved_spatial_weights = ucf101.definePath('notebook/201907/videoActionRecognitionStudy/out/checkpoints/spatial_hdf5/006-2.056.hdf5')
saved_temporal_weights = ucf101.definePath('notebook/201907/videoActionRecognitionStudy/out/checkpoints/temporal_hdf5/318-2.594.hdf5')
class_limit = None 
n_snip = 19 # number of chunks used for each video
opt_flow_len = 10 # number of optical flow frames used
image_shape=(224, 224)
original_image_shape=(341, 256)
batch_size = 8
fuse_method = 'average'

print(saved_spatial_weights)
print(saved_temporal_weights)
```

```python
print("class_limit = ", class_limit)

# Get the data.
data = DataSet(
        class_limit=class_limit,
        image_shape=image_shape,
        original_image_shape=original_image_shape,
        n_snip=n_snip,
        opt_flow_len=opt_flow_len,
        batch_size=batch_size,
        data_path=ucf101.getDataListPath(),
        opt_flow_path=ucf101.getOptFlowDir(),
        static_frame_path=ucf101.getTestDir()
        )
val_generator = data.validation_generator() # Get the validation generator
steps = data.n_batch

len(data.classes)

```

```python
# Get the model.
two_stream_fuse = ResearchModels(nb_classes=len(data.classes), n_snip=n_snip, opt_flow_len=opt_flow_len
                                 , image_shape=image_shape
                                 , saved_temporal_weights=saved_temporal_weights
                                 , saved_spatial_weights=saved_spatial_weights)

```

```python
# Evaluate!
myHistory = two_stream_fuse.model.fit_generator(generator=val_generator, steps_per_epoch=steps, max_queue_size=1)
```

```python
print('acc : ', myHistory.history['acc'])
print('top5_acc : ', myHistory.history['top_k_categorical_accuracy'])
```
